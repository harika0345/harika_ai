```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height = np.random.normal(loc=170, scale=10, size=100)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(height, kde=True)
plt.title("Height Distribution")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
height.mean()
```

```python
height.std()
```

```python
population = height[
    (height > (height.mean() - height.std())) &
    (height < (height.mean() + height.std()))
]
population
```

```python
population_percentage = (len(population) / len(height)) * 100
population_percentage
```

```python
tx_data = np.random.normal(loc=5000, scale=2000, size=1000)
tx_data[:10]

```

```python
tx_data = np.concatenate([tx_data, np.array([10000, 11000, 12000])])
```

```python
tx_data[-5:]
```

```python
mean = tx_data.mean()
mean
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(tx_data, kde=True)
plt.axvline(mean, color='r', linestyle='--')
plt.title("Transaction Amounts Distribution")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
outliers = tx_data[
    (tx_data < (mean - 3 * tx_data.std())) |
    (tx_data > (mean + 3 * tx_data.std()))
]
outliers
```
