1.Import NumPy

```python
import numpy as np
print("NumPy imported successfully")
```

2.Create NumPy Arrays

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([[10, 20, 30],
              [40, 50, 60]])

print("Array a:")
print(a)

print("\nArray b:")
print(b)
```

3.Shape, Dimension, Data Type

```python
print("\nShape of a:", a.shape)
print("Shape of b:", b.shape)

print("\nDimensions of a:", a.ndim)
print("Dimensions of b:", b.ndim)

print("\nData type of a:", a.dtype)
print("Data type of b:", b.dtype)
```

4.Random Data Generation

```python
rng = np.random.default_rng(42)

random_numbers = rng.integers(10, 100, size=10)

print("\nRandom Numbers:")
print(random_numbers)
```

5.Indexing & Slicing

```python
print("\nFirst element of a:", a[0])
print("Last element of a:", a[-1])

print("\nFirst row of b:")
print(b[0])

print("\nFirst column of b:")
print(b[:, 0])
```

6.Boolean Masking

```python
filtered = random_numbers[random_numbers > 50]

print("\nNumbers greater than 50:")
print(filtered)
#Important
```

7.Vectorized Operations

```python
print("\nArray a multiplied by 10:")
print(a * 10)

print("\nArray a plus 5:")
print(a + 5)
# No loops

```

8.Broadcasting

```python
broadcast_result = b + 100

print("\nBroadcasting example (b + 100):")
print(broadcast_result)
```

9.Aggregations & Statistics

```python
print("\nMean of a:", a.mean())
print("Sum of a:", a.sum())
print("Max of a:", a.max())
print("Index of max in a:", a.argmax())
```

10.Axis-based Aggregations

```python
print("\nColumn-wise mean of b:")
print(b.mean(axis=0))

print("\nRow-wise mean of b:")
print(b.mean(axis=1))
```

Linear Algebra Basics 

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5],
              [6]])

dot_product = A @ B

print("\nMatrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix multiplication (A @ B):")
print(dot_product)
```
