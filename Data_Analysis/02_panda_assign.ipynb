```python
# assignment
import numpy as np
import pandas as pd
# lets create the datafor retail domain with discount and other variable
data= {
    'product_id': np.random.randint(1000, 9999, 5000),
    'product_name': np.random.choice(['A', 'B', 'C', 'D', 'E'], 5000),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries'], 5000),
    'brand': np.random.choice(['P', 'Q', 'R'], 5000),
    'price': np.random.uniform(10, 100, 5000),
    'quantity': np.random.randint(1, 20, 5000),
    'discount': np.random.uniform(0, 0.3, 5000)
}

df = pd.DataFrame(data)
df.head(10)

# Questions

# 1. What is the average price of products in each category?
# 2. Which brand has the highest average discount?
# 3. What is the total revenue generated from each product?
# 4. How many products are there in each category?
# 5. What is the most popular product (highest quantity sold)?
```

# 1.Average price of products in each category

```python
avg_price_by_category = (df.groupby('category', as_index=False).agg(avg_price=('price', 'mean'))).sort_values(by='avg_price', ascending=False)

avg_price_by_category
```

#  2.Brand having highest average discount

```python
avg_discount_by_brand = (df.groupby('brand', as_index=False).agg(avg_discount=('discount', 'mean')).sort_values(by='avg_discount', ascending=False))

avg_discount_by_brand

```

# 3.Total revenue generated from each product

```python
df['revenue'] = df['price'] * df['quantity'] * (1 - df['discount'])

revenue_per_product = (df.groupby('product_id', as_index=False).agg(total_revenue=('revenue', 'sum')))

revenue_per_product.head()
```

# 4.Number of products in each category

```python
No_products_per_category = (df.groupby('category', as_index=False).agg(product_count=('product_id', 'count')))

No_products_per_category
```

# 5.Most popular product (highest quantity sold)

```python
popular_product = (df.groupby('product_name', as_index=False).agg(total_quantity_sold=('quantity', 'sum')).sort_values(by='total_quantity_sold', ascending=False))

popular_product
```

```python
import numpy as np
sales_data = {
  'cust_id': np.random.randint(1000, 9999, 5000),
  'age': np.random.randint(18, 70, 5000),
  'gender': np.random.choice(['M', 'F'], 5000),
  'location': np.random.choice(['New York', 'Los Angeles', 'Chicago'], 5000),
  'product': np.random.choice(['Laptop', 'Smartphone', 'Tablet'], 5000),
  'quantity': np.random.randint(1, 3, 5000)
}
sales_df = pd.DataFrame(sales_data)

# price_data
price_data = {
  'product': np.random.choice(['Laptop', 'Smartphone', 'Tablet'], 5000),
  'price': np.random.uniform(100, 1500, 5000)
}
price_df = pd.DataFrame(price_data)

# Questions which need to solve?

# 1. What is the average price of products sold to different age groups?
# 2. How does the gender distribution look across different product categories?
# 3. What is the most popular product in each location?
# 4. Are there any correlations between age and the price of products purchased? # ignore for now - we will discuss in our maths
# 5. How does the sales distribution look across different locations?
# 6. What is the average quantity of products sold per transaction?
# 7. What is the total revenue generated from each product category?
# 8. What is the average price of products sold to different age groups?
```

# Average price of products sold to different age groups 

```python
#merge tables 
final_table = sales_df.merge(price_df, on='product')
```

```python
# Age Group function 
def age_group(x):
    if x < 25:
        return '18-24'
    elif 25 <= x < 35:
        return '25-34'
    elif 35 <= x < 45:
        return '35-44'
    elif 45 <= x < 55:
        return '45-54'
    else:
        return '55+'


final_table['age_group'] = final_table['age'].apply(age_group)
```

```python
# Revenue (total price)
final_table['total_price'] = final_table['price'] * final_table['quantity']

final_table.head()
```

# Gender distribution across different categories 

```python
# 2. How does the gender distribution look across different product categories?
# lets do it from scratch
gender_distribution = final_table.groupby(['product', 'gender']).agg({'quantity': 'sum'})
# in terms of % age
def gender_percentage(x):
    return 100 * x / float(x.sum())

gender_distribution = gender_distribution.groupby(level=0).apply(gender_percentage)
gender_distribution
```

```python
# Step 1: aggregate quantity by product and gender
gender_distribution= (
    final_table
        .groupby(['product', 'gender']).agg({'quantity':'sum'})
)

# Step 2: calculate percentage within each product
gender_distribution['percentage'] = (
    gender_distribution
        .groupby('product')['quantity']
        .transform(lambda x: 100 * x / x.sum())
)

gender_distribution
```

# Most popular product in each location 

```python
product_location_sales = (
    final_table
        .groupby(['location', 'product'],as_index=False)
        .agg(total_quantity=('quantity', 'sum'))
)

product_location_sales
```

```python
most_popular_product = product_location_sales.loc[product_location_sales.groupby('location')['total_quantity'].idxmax()]
most_popular_product
```

# Sales distribution across different location 

```python
sales_distribution = (final_table.groupby('location', as_index=False).agg(total_sales=('quantity', 'sum')))

sales_distribution
```

```python
sales_distribution['sales_percentage']= 100 * sales_distribution['total_sales']/sales_distribution['total_sales'].sum()
sales_distribution
```

# average quantity of products sold per transaction

```python
avg_quantity_per_transaction = final_table.agg(avg_quantity_per_transaction=('quantity','mean'))

avg_quantity_per_transaction
```

```python
# Revenue (total price)
final_table['total_price'] = final_table['price'] * final_table['quantity']

```

```python
revenue_by_product = (
    final_table
        .groupby('product', as_index=False)
        .agg(total_revenue=('total_price', 'sum'))
)

revenue_by_product

```

```python
avg_price_by_age_group = (
    final_table
        .groupby('age_group', as_index=False)
        .agg(avg_price=('price', 'mean'))
)

avg_price_by_age_group
```
